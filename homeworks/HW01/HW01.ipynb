```python
import numpy as np
```

```python
# Создаем массив numpy разными способами

arr1 = np.array([8, 10, -3, 0.4])  # просто передаем массив чисел

temp_arr = [0, 4, -1]
arr2 = np.array(temp_arr)  # или можно передать уже созданный список

arr1, arr2
```

```python
# numpy также позволяет создавать матрицы

matrix = np.array([[3, 4], [5, 6], [7, 8]])
matrix
```

```python
# А еще можно воспользоваться встроенными функциями

arr3 = np.arange(5)         # перебирает числа от 0 до 5 и создает из них массив
arr4 = np.empty([3, 3])     # создает "пустой" массив; значения внутри - системный мусор
arr5 = np.zeros([2, 1])     # похоже на empty, только каждое значение инициализируется нулем

arr3, arr4, arr5
```

```python
# Посмотрим на форму и тип созданных массивов

print(f"arr1 shape: {arr1.shape} \
        \narr2 shape: {arr2.shape} \
        \nmatrix shape: {matrix.shape}")

print(f"arr1 type: {arr1.dtype} \
        \narr2 type: {arr2.dtype} \
        \nmatrix type: {matrix.dtype}")
```

Обратим внимание на то, что arr1 и arr2 имеет одномерную форму, когда как matrix - двумерную

Поработаем над матрицей

```python
# Можно указывать индексы как в двумерном массиве

print(matrix[0][0])  # левый верхний элемент
print(matrix[1][1])  # правый средний элемент
```

```python
# А можно через запятую

print(matrix[0, 0])
print(matrix[1, 1])
```

```python
# Теперь строки и столбцы

print(matrix[[0, 1], 1])    # берем первую и вторую строку, а значения из второго столбца
print(matrix[:,0])          # берем целиком первый столбец
```

Создадим две тестовых матрицы:

test1:\
(1 2 3 4\
5 6 7 8)

test2:\
(15 16 17 18\
-1 -2 -3 -4)

Обе имеют форму `(2, 4)`. По правилам линейной алгебры, такие массивы можно сложить поэлементно.

```python
test1 = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])
test2 = np.array([[15, 16, 17, 18], [-1, -2, -3, -4]])

print(test1 + test2)
```

Все верно! Однако что будет, если попробовать их перемножить? Для умножения матриц необходимо, чтобы число столбцов первой матрицы совпадало с количеством строк второй, что не подходит нашему случаю.

```python
print(test1 * test2)
```

Вместо ошибки произошло поэлементное умножение.

```python
# numpy предоставляет большое количество встроенных статистических функций

data = np.array([5, 9, -15, 4, 1, 0, 2, 10])

print(np.mean(data))    # среднее значение
print(np.median(data))  # медиана
print(np.var(data))     # дисперсия
print(np.std(data))     # стандартное отклонение
```
